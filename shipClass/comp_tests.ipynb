### checking the auto definition of transition matrices

```python
import numpy as np
import Component

# simulate 1000 components and determine the MTTF of the Markov Chains
num_comps = 100
steps = 500

failure_times = []

states = {0: "Working", 1: "Failed"}

for i in range(num_comps): 
    comp = Component.Component('Test_Component')
    comp.simulate(steps)
    
    print(comp.mC_model.history[-1])
    
    failure_times.append(comp.mC_model.failure_time)
    
print("Mean Time to Failure: ", np.mean(failure_times))


    
```
