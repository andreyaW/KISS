Simulating a ship with one system of multiple objects

```python
from shipClass.unmmanedShip import unmannedShip
from shipClass.System import System
from shipClass.SensedComp import SensedComp


# initializing a sytem of sensed components
sensed_comps = [SensedComp() for _ in range(5)]
system1 = System('sys1', sensed_comps)

sensed_comps2 = [SensedComp() for _ in range(4)]
system2 = System('sys2', sensed_comps2, [(1,2,3)])

# initializing an unmanned ship with the system of sensed components
ship = unmannedShip('ship1', [system1, system2])

ship.simulate(100)
ship.plotHistory()
ship.printHistory2Excel('history.xlsx')
```
